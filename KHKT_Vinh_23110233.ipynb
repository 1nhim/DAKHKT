```python
#5.1.1
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
    def display_info(self):
        print("Tên sách:", self.title)
        print("Tác giả:", self.author)
        print("Năm xuất bản:", self.year)
        print("---------------------")

book1 = Book("Dế Mèn Phiêu Lưu Ký", "Tô Hoài", 1941)
book2 = Book("Lão Hạc", "Nam Cao", 1943)


book1.display_info()
book2.display_info()
```

```python
#5.1.2
class BankAccount:
    def __init__(self, account_number, balance=0):
        self._account_number = account_number
        self._balance = balance
    @property
    def account_number(self):
        return self._account_number
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Nạp {amount} thành công.")
        else:
            print("Số tiền nạp không hợp lệ.")
    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Rút {amount} thành công.")
        else:
            print("Không đủ tiền hoặc số tiền không hợp lệ.")
    def get_balance(self):
        return self._balance


acc = BankAccount("123456789")
acc.deposit(1000)
acc.withdraw(300)
print("Số tài khoản:", acc.account_number)
print("Số dư hiện tại:", acc.get_balance())
```

```python
#5.1.3
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("Tên:", self.name)
        print("Tuổi:", self.age)
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def study(self):
        print(self.name, "đang học bài ")


sv = Student("Nguyễn Văn A", 20, "SV001")
sv.display_info()
print("Mã sinh viên:", sv.student_id)
sv.study()
```

```python
#5.1.4
class Vehicle:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        print("Hãng:", self.brand)
        print("Mẫu:", self.model)
        print("Năm:", self.year)
class Car(Vehicle):
    def __init__(self, brand, model, year, num_doors):
        super().__init__(brand, model, year)
        self.num_doors = num_doors

    def display_info(self):
        super().display_info()
        print("Số cửa:", self.num_doors)
        print("---------------------")
class Bike(Vehicle):
    def __init__(self, brand, model, year, type):
        super().__init__(brand, model, year)
        self.type = type

    def display_info(self):
        super().display_info()
        print("Loại xe:", self.type)
        print("---------------------")


car1 = Car("Toyota", "Camry", 2022, 4)
bike1 = Bike("Giant", "Escape 3", 2021, "Road")

car1.display_info()
bike1.display_info()
```

```python
#5.2.5

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


gen = (x for x in range(1, 21) if is_prime(x))


while True:
    try:
        print(next(gen))
    except StopIteration:
        break
```
