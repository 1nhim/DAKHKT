```python
def bai1_todo_list():
    """
    Cho phép người dùng nhập công việc, hiển thị danh sách
    có đánh số thứ tự và ghi vào file todo_list.txt.
    """
    print("\n" + "=" * 50)
    print("BÀI 1: ỨNG DỤNG TO-DO LIST")
    print("=" * 50)

    danh_sach = []

    print("Nhập các công việc cần làm (gõ 'xong' để kết thúc):")
    while True:
        cong_viec = input("  Công việc: ").strip()
        if cong_viec.lower() == "xong":
            break
        if cong_viec:  # Bỏ qua dòng trống
            danh_sach.append(cong_viec)

    if not danh_sach:
        print("Danh sách trống, không có gì để lưu.")
        return

    # Hiển thị danh sách ra màn hình
    print("\nDanh sách công việc của bạn:")
    for i, viec in enumerate(danh_sach, start=1):
        print(f"  {i}. {viec}")

    # Ghi đè toàn bộ danh sách vào file todo_list.txt
    with open("todo_list.txt", "w", encoding="utf-8") as f:
        for i, viec in enumerate(danh_sach, start=1):
            f.write(f"{i}. {viec}\n")

    print("\n✓ Đã lưu danh sách vào file 'todo_list.txt'.")

```

```python
import string

def tach_tu(van_ban):
    """Tách đoạn văn bản thành danh sách các từ, bỏ qua dấu câu cơ bản."""
    # Loại bỏ dấu câu cơ bản: dấu phẩy, dấu chấm
    for dau in [",", ".", "!", "?", ";", ":"]:
        van_ban = van_ban.replace(dau, "")
    danh_sach_tu = van_ban.split()
    return danh_sach_tu


def dem_tu(danh_sach_tu):
    """Đếm tổng số từ trong danh sách."""
    return len(danh_sach_tu)


def tim_tu_nhieu_nhat(danh_sach_tu):
    """Tìm từ xuất hiện nhiều nhất (không phân biệt hoa thường)."""
    if not danh_sach_tu:
        return None, 0

    # Chuyển tất cả về chữ thường để đếm
    tu_lower = [tu.lower() for tu in danh_sach_tu]

    # Đếm số lần xuất hiện của mỗi từ
    dem = {}
    for tu in tu_lower:
        dem[tu] = dem.get(tu, 0) + 1

    # Tìm từ có số lần xuất hiện cao nhất
    tu_nhieu_nhat = max(dem, key=dem.get)
    so_lan = dem[tu_nhieu_nhat]
    return tu_nhieu_nhat, so_lan


def bai2_phan_tich_van_ban():
    print("\n" + "=" * 50)
    print("BÀI 2: PHÂN TÍCH VÀ THỐNG KÊ VĂN BẢN")
    print("=" * 50)

    van_ban = input("Nhập đoạn văn bản (2-3 câu): ").strip()

    danh_sach_tu = tach_tu(van_ban)
    tong_so_tu = dem_tu(danh_sach_tu)
    tu_nhieu_nhat, so_lan = tim_tu_nhieu_nhat(danh_sach_tu)

    print(f"\nDanh sách các từ: {danh_sach_tu}")
    print(f"Tổng số từ       : {tong_so_tu}")
    print(f"Từ xuất hiện nhiều nhất: '{tu_nhieu_nhat}' ({so_lan} lần)")
```

```python
def tinh_bmi(can_nang, chieu_cao):
    """Tính chỉ số BMI."""
    return can_nang / (chieu_cao ** 2)


def phan_loai_bmi(bmi):
    """Phân loại kết quả BMI."""
    if bmi < 18.5:
        return "Gầy"
    elif bmi < 25:
        return "Bình thường"
    else:
        return "Thừa cân"


def bai3_tinh_bmi():
    print("\n" + "=" * 50)
    print("BÀI 3: CÔNG CỤ TÍNH CHỈ SỐ BMI THÔNG MINH")
    print("=" * 50)

    try:
        can_nang = float(input("Nhập cân nặng (kg): "))
        chieu_cao = float(input("Nhập chiều cao (m): "))

        bmi = tinh_bmi(can_nang, chieu_cao)
        phan_loai = phan_loai_bmi(bmi)

        print(f"\nChỉ số BMI của bạn: {bmi:.2f}")
        print(f"Phân loại         : {phan_loai}")

    except ValueError:
        print("Lỗi: Vui lòng nhập số, không nhập chữ!")
    except ZeroDivisionError:
        print("Lỗi: Chiều cao không được bằng 0!")
```

```python
import random
import time

def bai4_mini_hangman():
    print("\n" + "=" * 50)
    print("BÀI 4: TRÒ CHƠI ĐOÁN CHỮ CÁI (MINI HANGMAN)")
    print("=" * 50)

    danh_sach_tu = ["python", "apple", "river", "orange", "music", "school", "friend", "family", "flower", "mountain"]
    tu_bi_mat = random.choice(danh_sach_tu)

    so_luot_sai_toi_da = 10
    so_luot_sai = 0
    chu_da_doan = set()

    # Hiển thị ban đầu: dấu gạch ngang
    print(f"\nTừ bí mật có {len(tu_bi_mat)} chữ cái: {' '.join(['_'] * len(tu_bi_mat))}")

    thoi_gian_bat_dau = time.time()

    while so_luot_sai < so_luot_sai_toi_da:
        # Hiện trạng thái hiện tại
        trang_thai = " ".join(
            chu if chu in chu_da_doan else "_"
            for chu in tu_bi_mat
        )
        print(f"\nTrạng thái: {trang_thai}")
        print(f"Lượt sai còn lại: {so_luot_sai_toi_da - so_luot_sai}")
        print(f"Chữ đã đoán: {', '.join(sorted(chu_da_doan)) if chu_da_doan else 'Chưa có'}")

        # Kiểm tra thắng
        if all(chu in chu_da_doan for chu in tu_bi_mat):
            thoi_gian_ket_thuc = time.time()
            thoi_gian_hoan_thanh = round(thoi_gian_ket_thuc - thoi_gian_bat_dau, 2)
            print(f"\n🎉 Chúc mừng! Bạn đã đoán đúng từ '{tu_bi_mat}'!")
            print(f"Thời gian hoàn thành: {thoi_gian_hoan_thanh} giây")

            # Ghi vào file
            with open("hangman_result.txt", "w", encoding="utf-8") as f:
                f.write(f"Kết quả: Thắng\n")
                f.write(f"Từ bí mật: {tu_bi_mat}\n")
                f.write(f"Thời gian hoàn thành: {thoi_gian_hoan_thanh} giây\n")
                f.write(f"Số lượt sai: {so_luot_sai}\n")
            print("✓ Kết quả đã được lưu vào 'hangman_result.txt'.")
            return

        # Nhập chữ cái
        doan = input("Đoán một chữ cái: ").strip().lower()

        if len(doan) != 1 or not doan.isalpha():
            print("Vui lòng nhập đúng một chữ cái!")
            continue

        if doan in chu_da_doan:
            print(f"Bạn đã đoán chữ '{doan}' rồi!")
            continue

        chu_da_doan.add(doan)

        if doan in tu_bi_mat:
            print(f"✓ Đúng! Chữ '{doan}' có trong từ bí mật.")
        else:
            so_luot_sai += 1
            print(f"✗ Sai! Chữ '{doan}' không có trong từ bí mật.")

    # Thua
    print(f"\n💀 Bạn đã thua! Từ bí mật là '{tu_bi_mat}'.")
    with open("hangman_result.txt", "w", encoding="utf-8") as f:
        f.write(f"Kết quả: Thua\n")
        f.write(f"Từ bí mật: {tu_bi_mat}\n")
        f.write(f"Số lượt sai còn lại: 0\n")
    print("✓ Kết quả đã được lưu vào 'hangman_result.txt'.")
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'

def bai5_quan_ly_kho():
    print("\n" + "=" * 50)
    print("BÀI 5: QUẢN LÝ KHO HÀNG & BIỂU ĐỒ TỶ TRỌNG")
    print("=" * 50)

    # Dữ liệu tồn kho ban đầu
    kho_hang = {"Balo": 150, "Tui xach": 80, "Vali": 120}
    print(f"Kho hàng ban đầu: {kho_hang}")

    # Cập nhật thêm mặt hàng 'Vi da'
    so_luong_vi_da = int(input("\nNhập số lượng 'Vi da' muốn thêm vào kho: "))
    kho_hang["Vi da"] = so_luong_vi_da

    # Cập nhật 'Balo' giảm 30 đơn vị
    kho_hang["Balo"] -= 30

    print(f"\nKho hàng sau khi cập nhật: {kho_hang}")

    # Vẽ biểu đồ tròn (Pie chart)
    nhan = list(kho_hang.keys())
    gia_tri = list(kho_hang.values())

    mau_sac = ["#4e79a7", "#f28e2b", "#e15759", "#76b7b2"]

    fig, ax = plt.subplots(figsize=(7, 7))
    wedges, texts, autotexts = ax.pie(
        gia_tri,
        labels=nhan,
        autopct="%1.1f%%",
        colors=mau_sac,
        startangle=140,
        pctdistance=0.82,
        wedgeprops=dict(linewidth=1.5, edgecolor="white")
    )

    # Định dạng chữ
    for text in texts:
        text.set_fontsize(13)
    for autotext in autotexts:
        autotext.set_fontsize(11)
        autotext.set_fontweight("bold")

    ax.set_title("Tỷ Trọng Hàng Tồn Kho", fontsize=16, fontweight="bold", pad=20)
    plt.tight_layout()

    # Lưu biểu đồ
    ten_file = "bieu_do_kho_hang.png"
    plt.savefig(ten_file, dpi=150, bbox_inches="tight")
    print(f"\n✓ Biểu đồ đã được lưu vào file '{ten_file}'.")
    plt.show()
```
