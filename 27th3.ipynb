```python
#bai 1
def calculate_tax(income):
    if income < 5000:
        return 0
    elif income < 10000:
        return income * 0.1
    else:
        return income * 0.2


# Test cases
def test_calculate_tax():

    assert calculate_tax(4000) == 0


    assert calculate_tax(7000) == 700.0


    assert calculate_tax(12000) == 2400.0

    print("All tests passed!")


test_calculate_tax()
```

```python
#bai 2
def is_prime(n):
    if n <= 1: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


# Test cases
def test_is_prime():
    assert is_prime(1) == False
    assert is_prime(2) == True
    assert is_prime(3) == True
    assert is_prime(4) == False
    assert is_prime(17) == True
    assert is_prime(18) == False
    assert is_prime(19) == True

    print("All tests passed!")



test_is_prime()
```

```python
#bai 3
class User:
    def __init__(self, username):
        self.username = username

    def is_admin(self):
        return self.username == "admin"


# Test cases
def test_user_is_admin():

    user1 = User("admin")
    assert user1.is_admin() == True

    user2 = User("guest")
    assert user2.is_admin() == False

    user3 = User("user123")
    assert user3.is_admin() == False

    print("All tests passed!")


test_user_is_admin()
```

```python
#bai 4
from unittest.mock import patch

def send_welcome_email(email):
    print(f"Sending email to {email}")


def test_send_welcome_email():
    with patch("builtins.print") as mock_print:
        send_welcome_email("test@example.com")

        mock_print.assert_called_with("Sending email to test@example.com")

    print("Test passed!")


# Chạy test
test_send_welcome_email()
```

```python
#bai 5
from unittest.mock import patch, Mock
import requests

def fetch_user():
    response = requests.get("https://api.example.com/user")
    return response.json()


def test_fetch_user():
    mock_response = Mock()
    mock_response.json.return_value = {
        "id": 1,
        "name": "John Doe"
    }

    with patch("requests.get", return_value=mock_response) as mock_get:
        result = fetch_user()

        mock_get.assert_called_with("https://api.example.com/user")

        assert result == {
            "id": 1,
            "name": "John Doe"
        }

    print("Test passed!")


test_fetch_user()
```

```python
from unittest.mock import patch

def is_weekend():
    from datetime import datetime
    return datetime.now().weekday() >= 5


def test_is_weekend():
    with patch("datetime.datetime") as mock_datetime:
        mock_datetime.now.return_value.weekday.return_value = 5
        assert is_weekend() == True

    print("Test passed!")


test_is_weekend()
```

```python
#bai 7
def is_strong(password):
    return len(password) >= 8 and any(c.isdigit() for c in password)


def test_is_strong():
    # Password yếu
    assert is_strong("1234567") == False
    assert is_strong("abcdefgh") == False
    assert is_strong("abc123") == False

    # Password mạnh
    assert is_strong("abcd1234") == True
    assert is_strong("password1") == True
    assert is_strong("12345678") == True

    print("All tests passed!")


test_is_strong()
```

```python
#bai 8
def clean_input(s):
    return s.strip().lower().replace(" ", "_")


def test_clean_input():
    assert clean_input("  Hello  ") == "hello"

    assert clean_input("Hello World") == "hello_world"

    assert clean_input("  Python   Programming  ") == "python___programming"

    print("All tests passed!")


test_clean_input()
```

```python
#bai 9
def safe_divide(a, b):
    if b == 0:
        return None
    return a / b


def test_safe_divide():
    assert safe_divide(10, 2) == 5

    assert safe_divide(5, 0) is None

    assert safe_divide(-4, 2) == -2

    print("All tests passed!")


test_safe_divide()
```
